```python
### Week 8-1: Clustering : Coding exercise

## Exercise 1

import pandas

mower_df = pandas.read_csv("RidingMowers.csv")

X = mower_df[['Income','Lot_Size']]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

from scipy.cluster.hierarchy import dendrogram, linkage  

single = linkage(X_std,'single')
dendrogram(single)
```

```python
complete = linkage(X_std,'complete')
dendrogram(complete)
```

```python
average = linkage(X_std,'average')
dendrogram(average)
```
